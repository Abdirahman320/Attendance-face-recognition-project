```python
import matplotlib.pyplot as plt
import os

# Count images per student (each subfolder = 1 student)
dataset_path = 'data/preprocessed'
image_counts = [len(os.listdir(os.path.join(dataset_path, student)))
                for student in os.listdir(dataset_path)]

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(image_counts, bins=[0, 20, 40, 60, 80, 100, 150, 200], color='orange', edgecolor='black')
plt.title("Distribution of Image Counts per Student")
plt.xlabel("Number of Images")
plt.ylabel("Number of Students")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("histogram_distribution.png")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# True labels and predictions (same as before)
y_true = ['Ahmed', 'Ahmed', 'Amina', 'Mohamed', 'Amina',
          'Ahmed', 'Ahmed', 'Mohamed', 'Amina', 'Mohamed']

y_pred = ['Ahmed', 'Amina', 'Amina', 'Mohamed', 'Mohamed',
          'Ahmed', 'Mohamed', 'Mohamed', 'Amina', 'Ahmed']

# Generate confusion matrix
labels = ['Ahmed', 'Amina', 'Mohamed']
cm = confusion_matrix(y_true, y_pred, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)

# Plot
plt.figure(figsize=(6,5))
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix - Face Recognition Attendance")
plt.savefig("confusion_matrix_attendance.png")
plt.show()
```
